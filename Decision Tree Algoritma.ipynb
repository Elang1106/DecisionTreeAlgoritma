```
obat = pd.read_csv('drug200.csv')
```

```
d={'F' : 0, 'M' : 1}
```

```
obat['Sex'] = obat['Sex'].map(d)
```

```
d = {'LOW' : 0, 'NORMAL' : 1, 'HIGH' : 2}
```

```
obat['BP'] = obat['BP'].map(d)
```

```
d = {'NORMAL' : 0, 'HIGH' : 1}
```

```
obat['Cholesterol']=obat['Cholesterol'].map(d)
```

```
obat['Cholesterol']
```

```
obat.info()
```

```
fitur=['Age', 'Sex', 'BP', 'Cholesterol', 'Na_to_K']
```

```
X = obat[fitur]
```

```
y = obat['Drug']
```

```
from sklearn import tree
```

```
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
```

```
dtree = DecisionTreeClassifier()
dtree = dtree.fit(X,y)
```

```
tree.plot_tree(dtree, feature_names=fitur)
```
